```python

from transformers import pipeline

try:
    import torch
    print(f"PyTorch version: {torch.__version__}")
except ImportError:
    print("PyTorch is not installed. Please install it using 'pip install torch'")

# Check TensorFlow Installation
try:
    import tensorflow as tf
    print(f"TensorFlow version: {tf.__version__}")
except ImportError:
    print("TensorFlow is not installed. Please install it using 'pip install tensorflow'")

# Your existing notebook code

import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from transformers import pipeline
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Sample Data
documents = [
    "Sachin Tendulkar is one of the greatest batsmen in cricket history.",
    "The cricket match was interrupted by rain.",
    "The Indian Premier League (IPL), also known as the TATA IPL for sponsorship reasons, is a men's Twenty20 (T20) cricket league held annually in India. Founded by the BCCI (the Board of Control for Cricket in India) in 2007, the league features ten state or city-based franchise teams...",
    "Tendulkar has received several awards from the government of India: the Arjuna Award (1994), the Khel Ratna Award (1997), the Padma Shri (1998), and the Padma Vibhushan (2008)...",
    "The International Cricket Council (ICC) is the global governing body of cricket. It was founded as the Imperial Cricket Conference in 1909 by representatives from Australia, England, and South Africa...",
    "Board of Control for Cricket in India (BCCI) is the national governing body of cricket in India. Its headquarters is situated at the cricket centre in Churchgate, Mumbai. The BCCI is the wealthiest governing body of cricket in the world..."
]

# Preprocessing Function
def preprocess(text):
    # Tokenize
    tokens = word_tokenize(text.lower())
    
    # Remove stopwords and non-alphabetic tokens
    tokens = [word for word in tokens if word.isalpha() and word not in stopwords.words('english')]
    
    # Lemmatize
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    return ' '.join(tokens)

# Preprocess Documents
preprocessed_documents = [preprocess(doc) for doc in documents]

# TF-IDF Method
# Create a TF-IDF Vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the documents
tfidf_matrix = vectorizer.fit_transform(preprocessed_documents)

# Get feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Get the sum of TF-IDF scores for each word
tfidf_scores = tfidf_matrix.sum(axis=0).A1

# Create a dictionary of words and their scores
word_scores = dict(zip(feature_names, tfidf_scores))

# Sort the words by scores
sorted_word_scores = sorted(word_scores.items(), key=lambda x: x[1], reverse=True)

# Get the top N words (e.g., top 10 words)
top_words_tfidf = sorted_word_scores[:10]
print("Top words using TF-IDF:", top_words_tfidf)

# Load a pre-trained BERT model for feature extraction
nlp = pipeline("feature-extraction")

# Process each document to get embeddings
document_embeddings = [nlp(doc) for doc in preprocessed_documents]

# Example function to extract important words (simplified example)
def extract_important_words(doc_embedding, top_n=10):
    # Extracting important words based on the embedding can be more sophisticated
    # For simplicity, we return the first top_n tokens; a real implementation would
    # involve more advanced methods like attention weights, clustering, etc.
    # Here we simply print the shape of the embeddings as a placeholder
    return doc_embedding[0][:top_n]

# Extract top words from each document
important_words = [extract_important_words(doc_emb) for doc_emb in document_embeddings]
print("Important words using Transformers:", important_words)
```
